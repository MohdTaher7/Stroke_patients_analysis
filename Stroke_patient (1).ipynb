### Check for missing values

```python
#Display the count of missing values for each column
df.isnull().sum()
```

As you can see, bmi column has 201 null value appeared in the data frame; those missing values may hinder further analysis. 
<div>So, how do we identify all those missing values and deal with them?</div> 


<b>How to work with missing data?</b>

Steps for working with missing data:
<ol>
    <li>Identify missing data</li>
    <li>Deal with missing data</li>
    <li>Correct data format</li>
</ol>


```python
# Calculate the missing values percentage for each column and round to two decimal places
missing_values_percentage = (df.isnull().mean() * 100).round(2)

# Display the missing values percentage for each column
print("Missing Values Percentage:\n")
print(missing_values_percentage)
```

**Observation:**
* We can see that **around 3.93%** of the "bmi" columns value are missing and hence we cannot drop this much data. We will fill these columns with "Unknown"


<h4>Evaluating for Missing Data</h4>

The missing values are converted by default. Use the following functions to identify these missing values. You can use two methods to detect missing data:
<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.


```python
df['bmi'].isnull().any
```

### Deal with missing data
<b>How should you deal with missing data?</b>

<ol>
    <li>Drop data<br>
        a. Drop the whole row<br>
        b. Drop the whole column
    </li>
    <li>Replace data<br>
        a. Replace it by mean<br>
        b. Replace it by frequency<br>
        c. Replace it based on other functions
    </li>
</ol>


## We can use any one of method and I am using Mean 

```python
#  mean of bmi column
df['bmi'].fillna(0) 
np.mean(df.bmi)
```

```python
#  mean is to replace with zeroes
df.fillna(np.mean(df.bmi), inplace = True)
np.mean(df.bmi)
```

```python
# dataset after cleaning
df
```

<b>Good!</b> Now, we have a data set with no missing values.
